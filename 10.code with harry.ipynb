## **decorators**

```python
#one function that change and update an another function
#modify an funtion
# syntax:@decoder function
# def my_function
#  pass
```

```python
def hello():
  print("hello world")
hello()
```

```python
def greet(fx):
  def mfx():
    print("good morning")
    fx()
    print("thanks for using this function")
  return mfx
@greet  #syntax
def hello():
  print("hello world")
hello()
```

## **or**


```python
def greet(fx):
  def mfx():
    print("good morning")
    # fx()
    print("thanks for using this function")
  return mfx
 #syntax
def hello():
  print("hello world")
greet(hello)()
```

```python
def greet(fx):
  def mfx(*args,**kwargs):
    print("good morning")
    fx(*args,**kwargs)
    print("thanks for using this function")
  return mfx
@greet  #syntax
def hello():
  print("hello world")
def add(a,b):
  print(a+b)
hello()
greet(add)(1,2)

```

## **getters and setters**

```python
class Myclass:
  def __init__(self,value):
    self._value=value
  def show(self):
    print(f"value is {self._value}")
  @property
  def value(self):
    return self._value
obj=Myclass(10)
print(obj._value)
obj.show()
```

```python
class Myclass:
  def __init__(self,value):
    self._value=value
  def show(self):
    print(f"value is {self._value}")
  @property
  def ten_value(self):
    return self._value
  @ten_value.setter
  def value(self,new_value):
    self._value=new_value/10
    return self._value
obj=Myclass(10)
print(obj._value)
obj.show()
```
