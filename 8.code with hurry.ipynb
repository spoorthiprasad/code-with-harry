```python
#lambda function  #((argument:expression))
def double(x):
  return x*2
print(double(5))

```

```python
double=lambda x:x*2   #instead of define function we use lambda function
print(double(5))
```

```python
double=lambda x:x*x
cube=lambda x:x*x*x
print(double(3))
print(cube(3))
```

```python
avg=lambda x,y:(x+y)/2
print(avg(5,9))
```

```python
#map filter and reduce
def cube(x):
  return x*x*x
print(cube(5))
```

```python
def cube(x):
  return x*x*x
print(cube(5))
l=[1,2,3,4,5,6,7]  #map(function ,iteration)
newl=list(map(cube,l))#it is used as a higher function here function passe an function as a argument
print(newl)
```

```python
def cube(x):
  return x*x*x
print(cube(5))
l=[1,2,3,4,5,6,7]
newl=[]
for item in l:
  newl.append(cube(item))
print(newl)
```

```python
l=[1,2,3,4,5,6,7]
def filter_function(a):
  return a>4
newnewl=list(filter(filter_function,l))   #filter_function:here it cinsider l as true or false then it send that to newnewl
newnewl
```

```python
l=[1,2,3,4,5,6,7]
newnewl=list(map(lambda x:x*x*x,l))
newnewl
```

```python
#reduce
from functools import reduce
l=[1,2,3,4,5,6,7]
def  mysum(x,y):
  return x+y
sum=reduce(mysum,l)
print(sum)
```

```python
from functools import reduce
l=[1,2,3,4,5,6,7]
sum=reduce(lambda x,y:x + y,l)
print(sum)
```
