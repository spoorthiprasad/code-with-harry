## **Enumerate function in python**

```python
marks=[25,45,87,78,89,90,100,1,8,9]
index=0
for mark in marks:
  print(mark)
  if(index==3):
    print("awesome spoo")
  index+=1
```

```python
marks=[25,45,87,78,89,90,100,1,8,9]
index=0
for index, mark in enumerate(marks):   #both index and value will applied at a same time
  print(mark)
  if(index==3):
    print("awesome spoo")
```

```python
marks=[25,45,87,78,89,90,100,1,8,9]
index=0
for index, mark in enumerate(marks,start=1):   #always strt from 0 here we intialize it to 1
  print(mark)
  if(index==3):
    print("awesome spoo")
```
